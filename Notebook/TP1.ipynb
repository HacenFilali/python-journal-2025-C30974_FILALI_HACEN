## Info() Method &  Sample() Method

```python
df.info() # provides a full information of the entire DataFrame
print(df.info())
df.info(verbose=False) # provides a short information of the entire DataFrame
print(df.info(verbose=False))
df.sample(5) # In this sample, the function returns 5 random subsets
print(df.sample(5))
df.sample(frac = 0.1) # In this sample, the function returns a fraction of randot
print(df.sample(frac=0.1))
```
